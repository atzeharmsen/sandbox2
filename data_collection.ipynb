```python
# Make the heatmap
# assumption: above the highest line it stays constant

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Add cornerpoints to fill above value = 1.0 line
added_cornerpoints = [
    {"latitude": -87.0, "altitude": 60, "value": 1.0},
    {"latitude": 87.0, "altitude": 60, "value": 1.0}
]
b_df = pd.concat([b_df, pd.DataFrame(added_cornerpoints)], ignore_index=True)

# Extract columns
x = b_df['latitude']
y = b_df['altitude']
z = b_df["value"].astype(float).values  # z: value

# Create grid
xi = np.linspace(-85, 86, 171)
yi = np.linspace(0, 60, 601)
X, Y = np.meshgrid(xi, yi)

# Interpolate values onto grid
Z = griddata((x, y), z, (X, Y), method='linear')
# Z[Z > 1] = 1
# Plot
plt.figure(figsize=(10, 6))
heatmap = plt.pcolormesh(X, Y, Z, shading='auto', cmap='viridis')
plt.colorbar(heatmap, label='Concentration (ppmv)')

plt.xlabel('Latitude (deg)')
plt.ylabel('Altitude (km)')
plt.title('SWV ppmv due to CH4 change')
plt.tight_layout()
plt.scatter(b_df[b_df["source"] == "b10"]["latitude"], b_df[b_df["source"] == "b10"]["altitude"], color="red", label="b10")

plt.show()
```

```python
from ambiance import Atmosphere
N_avogrado = 6.022*10**23
M_h2o = 18*10**-3
delta_h = 100.
delta_deg = 1.
heights = yi#np.arange(0, 60000 + delta_h, delta_h)  # 0 to 60 km
latitudes = xi#np.arange(-85, 86, delta_deg)  # -85° to 85°

number_density = Atmosphere(yi).number_density
R = 6371000.  # Earth radius in meters
delta_phi = np.deg2rad(delta_deg)
# Volume of 1° latitude x 100 m height strip integrated over all longitude
volumes = np.zeros((len(heights), len(latitudes)))
for i, h in enumerate(heights):
    for j, lat in enumerate(latitudes):
        volumes[i, j] = 2 * np.pi * (R + h) ** 2 * np.cos(np.deg2rad(lat)) * delta_phi * delta_h

parts_mat = volumes * number_density[:, np.newaxis]

SWV_mol_mat = parts_mat * Z * 10 ** -6 / N_avogrado
SWV_mass_mat = SWV_mol_mat * M_h2o


plt.figure(figsize=(10, 6))
heatmap = plt.pcolormesh(X, Y, SWV_mass_mat, shading='auto', cmap='viridis')
plt.colorbar(heatmap, label='Concentration (kg)')

plt.xlabel('Latitude (deg)')
plt.ylabel('Altitude (km)')
plt.title('SWV kg due to CH4 change')
plt.tight_layout()
# plt.scatter(b_df[b_df["source"] == "b10"]["latitude"], b_df[b_df["source"] == "b10"]["altitude"], color="red", label="b10")

plt.show()
```
