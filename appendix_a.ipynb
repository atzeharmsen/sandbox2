```python
import numpy as np
warming_2019_2050 = 0.06 #K
nowack = 0.31 #ppmv/K
period = 2050-2019

h2o_entry_increase = warming_2019_2050 * nowack # ppmv
h2o_increase_per_year = h2o_entry_increase /period
def ppmv_to_Tg_per_Tg(conc):
    Mswv = 18.01528 * 10**-3  # kg/mol
    Mair = 28.97 * 10**-3  # kg/mol
    mass_ratio = conc*10**-6*Mswv/Mair
    return mass_ratio

mass_entering = np.array([6*10**8,9*10**8])
additional = ppmv_to_Tg_per_Tg(h2o_increase_per_year)*mass_entering
total_additional = additional*period
print(additional)
print(total_additional)
```
