```python
import xarray as xr
from matplotlib import pyplot as plt

data_nc = r"C:\Users\atzeh\PycharmProjects\OAC_Thesis\ATZE\dissertation_fig_7-9.nc"
data = xr.load_dataset(data_nc)
data
```

```python
tau = data["tau_H2O"]            # your data variable
alt = data["Altitude"]
lat = data["Latitude"]
from matplotlib.colors import LogNorm
plt.figure(figsize=(8,5))
# Make a contour/heatmap
plt.pcolormesh(lat, alt, tau, shading="auto", cmap="viridis", norm=LogNorm())   # <--- log color scale)
plt.colorbar(label="tau")
plt.xlabel("Latitude")
plt.ylabel("Altitude")
plt.title("Tau vs Altitude & Latitude")
plt.show()
```
