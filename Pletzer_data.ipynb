```python
import numpy as np
from scipy.interpolate import RegularGridInterpolator
import matplotlib.pyplot as plt

# Example small dataset
small_data = np.array(data['tau_H2O'])#np.random.rand(6, 8)

# Original axes
x_small = np.array(data['Latitude'])+90  # Linear axis
y_small = np.array(data['Altitude'])    # Logarithmic axis

# Define target axes (same size as large dataset)
x_large = np.unique(data2['lat'].values)   # Linear axis, 41 points
y_large = np.unique(data2['plev'].values)   # Logarithmic axis, 151 points

# Create interpolator
interpolator = RegularGridInterpolator(
    (y_small, x_small),  # Note: order is (rows, columns) = (y, x)
    small_data,
    method='linear',
    bounds_error=False,  # Do not throw error for out-of-bounds
    fill_value=None  
)

# Create meshgrid for target points
Y_large, X_large = np.meshgrid(y_large, x_large, indexing='ij')

# Flatten points for interpolator
points = np.array([Y_large.ravel(), X_large.ravel()]).T

# Interpolate
large_data = interpolator(points).reshape(len(y_large), len(x_large))

print("Original small data shape:", small_data.shape)
print("Interpolated large data shape:", large_data.shape)

# Optional: visualize
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(small_data, aspect='auto', origin='lower', extent=[x_small[0], x_small[-1], y_small[0], y_small[-1]],)
plt.gca().invert_yaxis() 
plt.yscale('log')          # Set y-axis to logarithmic scale
plt.colorbar(label="tau")

plt.title('Original Small Data')

plt.subplot(1,2,2)
plt.imshow(large_data, aspect='auto', origin='lower', extent=[x_small[0], x_small[-1], y_small[0], y_small[-1]],)
plt.gca().invert_yaxis() 
plt.yscale('log')          # Set y-axis to logarithmic scale

plt.title('Interpolated Large Data')
plt.colorbar(label="tau")

plt.show()
```
