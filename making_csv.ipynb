```python
import xarray as xr
import numpy as np

def save_myhre_dataframe_to_netcdf2(df, filename):
    """
    Save dataframe (latitude, altitude, value, source) to NetCDF.
    """

    ds = xr.Dataset(
        {
            "value": ("point", df["value"].values),
            "source": ("point", df["source"].astype(str).values),
        },
        coords={
            "latitude": ("point", df["latitude"].values),
            "altitude": ("point", df["altitude"].values),
        }
    )

    # Metadata
    ds.latitude.attrs["units"] = "degrees_north"
    ds.altitude.attrs["units"] = "m"
    ds.value.attrs["units"] = "ppm"
    ds.attrs["Title"] = "HALOE zonal mean vertical profile of CH4 over the period 1991 throughout 1999"
    ds.attrs["Author"] = "Atze Harmsen"
    ds.attrs["Contact"] = "atzeharmsen@gmail.com"
    ds.attrs["Reference"] = "Myhre et al. (2007), Stratospheric water vapour from CH4; Figure 1"

    ds.to_netcdf(filename)

    print(f"Saved to {filename}")
```
