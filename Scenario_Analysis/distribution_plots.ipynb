```python
import openairclim as oac
from openairclim.calc_ch4 import calc_swv_mass_conc
import numpy as np
time = np.arange(2010, 2100, 1)
REV_ch4_conc = 1800*np.ones(len(time))
S2_ch4_conc = 2*REV_ch4_conc
S5_ch4_conc = 5*REV_ch4_conc

mass_REV,conc_REV,distr_REV = calc_swv_mass_conc(REV_ch4_conc)
mass_S2,conc_S2,distr_S2 = calc_swv_mass_conc(S2_ch4_conc)
mass_S5,conc_S5,distr_S5 = calc_swv_mass_conc(S5_ch4_conc)
```

```python
import matplotlib.pyplot as plt
from ambiance import Atmosphere

# Define constants
delta_h = 100.0  # height increment in meters
delta_deg = 1.0  # latitude increment
heights = np.arange(0, 60000 + delta_h, delta_h)  # 0 to 60 km

latitudes = np.arange(
    -85, 85, delta_deg)
        
        
plt.figure(figsize=(10, 6))
# contour_levels = np.arange(0.0, 1.1, 0.1)  # adjust levels
contours = plt.contour(
    latitudes,
    Atmosphere(heights).pressure / 100,
    distr_REV/1000,
    # levels=contour_levels,
    colors="k",
    linewidths=0.8,
)
plt.xlim([-90, 90])
plt.ylim([1, 1000])

# --- Add contour labels ---
plt.clabel(contours, inline=True, fmt="%.1f", fontsize=8)
plt.yscale("log")
plt.gca().invert_yaxis()  # invert so low pressure is at the top
plt.xlabel("Latitude (deg)")
plt.ylabel("Altitude (hPa)")
plt.title("SWV concentration in ppmv")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
# contour_levels = np.arange(0.0, 1.1, 0.1)  # adjust levels
contours = plt.contour(
    latitudes,
    Atmosphere(heights).pressure / 100,
    distr_S2/1000,
    # levels=contour_levels,
    colors="k",
    linewidths=0.8,
)
plt.xlim([-90, 90])
plt.ylim([1, 1000])

# --- Add contour labels ---
plt.clabel(contours, inline=True, fmt="%.1f", fontsize=8)
plt.yscale("log")
plt.gca().invert_yaxis()  # invert so low pressure is at the top
plt.xlabel("Latitude (deg)")
plt.ylabel("Altitude (hPa)")
plt.title("SWV concentration in ppmv")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
# contour_levels = np.arange(0.0, 1.1, 0.1)  # adjust levels
contours = plt.contour(
    latitudes,
    Atmosphere(heights).pressure / 100,
    distr_S5/1000,
    # levels=contour_levels,
    colors="k",
    linewidths=0.8,
)
plt.xlim([-90, 90])
plt.ylim([1, 1000])

# --- Add contour labels ---
plt.clabel(contours, inline=True, fmt="%.1f", fontsize=8)
plt.yscale("log")
plt.gca().invert_yaxis()  # invert so low pressure is at the top
plt.xlabel("Latitude (deg)")
plt.ylabel("Altitude (hPa)")
plt.title("SWV concentration in ppmv")
plt.tight_layout()
plt.show()
```

```python
# Calculate the percentage increase

p2 = (distr_S2-distr_REV)/distr_REV*100
p5 = (distr_S5-distr_REV)/distr_REV*100


plt.figure(figsize=(10, 6))
# X, Y = (len(grid[0, :]), len(grid[:, 0]))
heatmap = plt.pcolormesh(
    latitudes,
    Atmosphere(heights).pressure / 100,  # make it hPa
    p2,
    shading="auto",
    cmap="viridis",
)
plt.colorbar(heatmap, label="Percent")
plt.yscale("log")
plt.gca().invert_yaxis()  # invert so low pressure is at the top
plt.xlabel("Latitude (deg)")
plt.ylabel("Altitude (hPa)")
plt.title("increase in SWV for S2")
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
# X, Y = (len(grid[0, :]), len(grid[:, 0]))
heatmap = plt.pcolormesh(
    latitudes,
    Atmosphere(heights).pressure / 100,  # make it hPa
    p5,
    shading="auto",
    cmap="viridis",
)
plt.colorbar(heatmap, label="Percent")
plt.yscale("log")
plt.gca().invert_yaxis()  # invert so low pressure is at the top
plt.xlabel("Latitude (deg)")
plt.ylabel("Altitude (hPa)")
plt.title("increase in SWV for S5")
plt.tight_layout()
plt.show()
```
